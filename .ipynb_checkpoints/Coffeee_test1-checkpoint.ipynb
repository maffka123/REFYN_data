```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from scipy.optimize import curve_fit

sns.set()
```

## Explore .csv files

We have 2 types of data:

* Point data (James_Daily_Coffee.csv)

* 1 Timeseries (Coffee_Data_Day2.csv)

### Point data

#### Simple import

```python
j_coffee = pd.read_csv('Desktop/James_Daily_Coffee.csv')
j_coffee.head()
```

#### Adjust data types and column content

```python
j_coffee = j_coffee.drop(columns = ['Unnamed: 11'])
```

```python
j_coffee = j_coffee.rename(columns = {'Air Temp':'Air Temp, °C', 'Humidity Reading':'Humidity Reading, %', 
                                      'Date ':'Date', 'Time of day ':'Time of day'})
j_coffee['Air Temp, °C'] = j_coffee['Air Temp, °C'].apply(lambda x: x.replace('°C', ''))
j_coffee['Humidity Reading, %'] = j_coffee['Humidity Reading, %'].apply(lambda x: x.replace('%', ''))
j_coffee['Grind Setting'] = j_coffee['Grind Setting'].apply(lambda x: x.lower().replace('turco', ''))
j_coffee['Grind Setting'] = pd.to_numeric(j_coffee['Grind Setting'])
j_coffee['Humidity Reading, %'] = pd.to_numeric(j_coffee['Humidity Reading, %'])
j_coffee['Air Temp, °C'] = pd.to_numeric(j_coffee['Air Temp, °C'])
```

```python
j_coffee['Full_Date'] = j_coffee[['Date', 'Time of day']].apply(lambda x: ' '.join(x), axis=1)
j_coffee['Full_Date'] = pd.to_datetime(j_coffee['Full_Date'])
j_coffee['Hour'] = j_coffee['Full_Date'].dt.hour
j_coffee['Minute'] = j_coffee['Full_Date'].dt.minute
j_coffee = j_coffee.drop(columns = ['Date', 'Time of day', 'Notes'])
j_coffee
```

#### Take a look at Notre Café n°26

```python
j_coffee_n26 = j_coffee[j_coffee['Bean Name'] == 'Notre Café n°26']
j_coffee_n26 = j_coffee_n26[['Grind Setting','First Drip Time', 'Pour Time', 'Dose in',
                               'Humidity Reading, %', 'Air Temp, °C', 'Hour', 'Minute', 'Coffee out']]
j_coffee_n26_x = j_coffee_n26[['Grind Setting','First Drip Time', 'Pour Time', 'Dose in',
                               'Humidity Reading, %', 'Air Temp, °C', 'Hour', 'Minute']]

j_coffee_n26_y = j_coffee_n26[['Coffee out']]
```

It is always good to see correlations

```python
sns.pairplot(j_coffee_n26)
```

```python
corr = j_coffee_n26.corr()
mask = np.triu(np.ones_like(corr, dtype=np.bool))
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

Simple linear regression

```python
predicts = []
for i in range(len(j_coffee_n26)):
    j_coffee_n26_x_train = j_coffee_n26_x.drop(j_coffee_n26_x.index[[i]])
    j_coffee_n26_y_train = j_coffee_n26_y.drop(j_coffee_n26_x.index[[i]])
    regr = linear_model.LinearRegression()
    regr.fit(j_coffee_n26_x_train, j_coffee_n26_y_train)
    y = regr.predict(j_coffee_n26_x.iloc[i,:].values.reshape(1, -1))
    predicts.append(y[0][0])
    
```

```python
plt.scatter(j_coffee_n26_y['Coffee out'].values.tolist(), predicts)
plt.xlabel('True Coffee out')
plt.ylabel('Predicted Coffee out')
plt.xlim((24,60))
plt.ylim((24,60))
```

### Timeseries

#### Simple import

```python
day_coffee = pd.read_csv('Desktop/Coffee_Data_Day2.csv')
day_coffee.head()
```

#### Adjust data types and column content

```python
day_coffee = day_coffee.rename(columns = {'Temp':'Air Temp, °C', 'Humidity Reading':'Humidity Reading, %'})

day_coffee['Humidity Reading, %'] = day_coffee['Humidity Reading, %'].apply(lambda x: x.replace('%', '').replace(',', '.'))
day_coffee['Air Temp, °C'] = day_coffee['Air Temp, °C'].apply(lambda x: x.replace(',', '.'))
day_coffee['Coffee out'] = day_coffee['Coffee out'].apply(lambda x: x.lower().replace(',', '.'))
day_coffee['Dose in'] = day_coffee['Dose in'].apply(lambda x: x.lower().replace(',', '.'))

day_coffee['Grind Setting'] = day_coffee['Grind Setting'].apply(lambda x: x.lower().replace('turco', ''))


day_coffee['Grind Setting'] = pd.to_numeric(day_coffee['Grind Setting'])
day_coffee['Humidity Reading, %'] = pd.to_numeric(day_coffee['Humidity Reading, %'])
day_coffee['Air Temp, °C'] = pd.to_numeric(day_coffee['Air Temp, °C'])
day_coffee['Coffee out'] = pd.to_numeric(day_coffee['Coffee out'])
day_coffee['Dose in'] = pd.to_numeric(day_coffee['Dose in'])

day_coffee.head()
```

```python
day_coffee['Time'] = day_coffee['Time'].apply(lambda x: '08.06.2020 '+x)
day_coffee['Time'] = pd.to_datetime(day_coffee['Time'])
#day_coffee['timedela_with_previous_step'] = 
```

### Take a look at the timeseries

```python
fig, ax = plt.subplots(figsize=(15,5))
g = sns.lineplot(x = 'Time', y = 'Coffee out', data = day_coffee, ax = ax)
plt.setp(g.get_xticklabels(), rotation=30)
plt.show()
```

```python
def heating_index(timeseries,a,b):
    index = a
    index_max = 1
    for i in range(len(timeseries)-1):
        dt = (timeseries[i+1] - timeseries[i]) / np.timedelta64(1, 'm')
        index += a - np.exp(b*dt)
        
    return(index)
```

```python
popt, pcov = curve_fit(heating_index, day_coffee['Time'].values, day_coffee['Coffee out'].values)
popt
```

```python
index = []
index.append(0)
for i in range(2,len(day_coffee)+1):
    timeseries = day_coffee['Time'].values[:i]
    index.append(heating_index(timeseries,3.80663726e+08, 1.82084431e-01))
    
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(day_coffee['Time'].values, index)
```

```python
delta = (day_coffee['Time']-day_coffee['Time'].shift()).fillna(0)
delta_x = np.cumsum(delta.values / np.timedelta64(1, 'm'))

def heating_index2(timeseries,a,b):
    index = 1
    index_max = 1
    for i in range(len(timeseries)-1):
        dt = timeseries[i+1] - timeseries[i]
        index += 36.2*(a - np.exp(-b*dt))
        
    return(index)

popt, pcov = curve_fit(heating_index2, delta_x, day_coffee['Coffee out'].values)
popt
```

```python
index = []
index.append(36.2)
for i in range(2,len(day_coffee)+1):
    timeseries = delta_x[:i]
    index.append(heating_index2(timeseries,popt[0], popt[1]))
    
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(delta_x, index)
```
